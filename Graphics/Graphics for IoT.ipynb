```python
import matplotlib.pyplot as plt
import numpy as np

def leggi_file(nome_file):
    dati = []
    with open(nome_file, 'r') as file:
        for riga in file:
            dato = float(riga.strip())
            dati.append(dato)
    return dati

def crea_grafico(dati1, dati2, dati3):
    plt.plot(dati1, label='50 cluster')
    plt.plot(dati2, label='10 cluster')
    plt.plot(dati3, label='3 cluster')
    plt.xlabel('Tempo [#TimeSlot]')
    plt.ylabel('Ritardo medio [#TimeSlot]')
    plt.title('Grafico del ritardo medio')
    plt.grid(True)
    plt.legend()
    plt.show()

data1 = leggi_file("50.txt")
data2 = leggi_file("10.txt")
data3 = leggi_file("3.txt")

window = 5

average_data_1 = []
for ind in range(len(data1) - window + 1):
    average_data_1.append(np.mean(data1[ind:ind+window]))

average_data_2 = []
for ind in range(len(data2) - window + 1):
    average_data_2.append(np.mean(data2[ind:ind+window]))

average_data_3 = []
for ind in range(len(data3) - window + 1):
    average_data_3.append(np.mean(data3[ind:ind+window]))

crea_grafico(average_data_1, average_data_2, average_data_3)
```
